```python
#importing Relevant Libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
url="https://en.wikipedia.org/wiki/Narendra_Modi"# url of the webpage

r=requests.get(url)# GET requesting the url
r.status_code # checking the status of get request
```

```python
htmlContent=r.content # loading the content 
htmlContent
```

```python
soup=BeautifulSoup(htmlContent,"html.parser") # instantstiating the BeautifulSoup Class
print(soup)
```

```python
bio=soup.find_all("tbody")[0]
```

As,the Data required for us is in the First tbody

```python
#Below line of code will Retrieve throughout 'th' tags in bio and return data in text format
raw_info=[i.text for i in bio.find_all('th')]

#len(raw_info)->38

raw_info
```

```python
info=raw_info[30:38] # -> Taking only the info Required
print(info)
```

```python
#Below line of code will Retrieve throughout 'td' tags in bio and return data in text format after striping whitespaces
raw_details=[i.text.strip() for i in bio.find_all("td")]

#len(raw_details)->49

raw_details
```

```python
details=raw_details[-10:-2] # -> Taking only the details Required
details
```

```python
personal_info=pd.DataFrame(details,index=info,columns=["Personal info PM MODI"])
personal_info
```
